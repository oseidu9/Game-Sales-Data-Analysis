```python
#import needed libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#read in the data
#read_csv is pandas command. r needed before directory for RAW directory path reading
#result of df is essentially the csv file converted to a pandas table.
df = pd.read_csv(r'D:\Soba\Downloads\Video_Games_Sales_as_at_22_Dec_2016.csv')

#df stands for dataframe. It is basically a table.
```

```python
#look at the data that has been read/imported
#note that some fields will be NaN which is NULL. because it has no values there
df.head()
```

```python
#explicitly check for percentage of each column where value is null 
for col in df.columns:
    
    #np is alias for numpy library. np.mean() is command that finds mean
    #df[col].isnull() returns true or false for each entry in that column if it is null or not
    #with np.mean(), if it takes in a boolean, True = 1 and False = 0
    #so it is essentially sum of all boolean values for that column/number of entries in column
    #giving us percentage of entries that have null.
    pct_missing = np.mean(df[col].isnull())
    
    #also, format how you want the above to be displayed.
    #first variable goes in first curly brace, second in second. and rest are concatenated together
    print('{} - {}%'.format(col, pct_missing*100))
```

```python
#some float columns have .0 at the end, but not needed. Remove them (small data clean)

#change data type to int for the columns that have this issue
df['Year_of_Release'] = df['Year_of_Release'].astype('Float64')
df['Critic_Score'] = df['Critic_Score'].astype('Float64')

#before converting User_Score to float, it has a tbd field (non-numeric) for No User Score. Convert whilst nulling these.
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce', downcast='float')
df['Critic_Count'] = df['Critic_Count'].astype('Float64')
df['User_Count'] = df['User_Count'].astype('Float64')

#NOTE: since some of these fields have null values, you must use Int64, not int64, otherwise errors
```

```python
#Dropping duplicates from table

#the below will show us values from this column that are unique in descending order
#No removal done, it just shows us the output of below
#df['Name', 'Platform'].drop_duplicates().sort_values(ascending=False)

#to ACTUALLY do the duplicate removal, set the Name column to equal the above code.
#df['Name', 'Platform'] = df['Name', 'Platform'].drop_duplicates().sort_values(ascending=False)

#BUT this only removes the columns specified. if you want to remove whole tuple duplicates, do:
#df.drop_duplicates()

#BUT what we want to do is remove tuples where the name and platform combination already exists.
#can have a temp hashhmap. Key = concatenation between name and platform.
#just set value to 1. If key exists, remove tuple from the table. Identify it by row number.

namePlatform = {}
number = 0
for index, row in df.iterrows():
    #if namePlatform key does not exist, add it to namePlatform dict, else remove row from table
    if str(row['Name']) + str(row['Platform']) not in namePlatform:
        namePlatform[str(row['Name']) + str(row['Platform'])] = 1
    else:
        
        #first arg means remove this row of this index, axis=0 means delete row from dataframe,
        #inplace=True means drop the row from dataframe (so no null rows i guess?)
        df.drop([index], axis=0, inplace=True)
        
        
        
        
```

```python
#scatter plot of critic score vs sales (BUT use a cleaned version that omits rows with NULL values in specified columns.).
df_cleaned = df.dropna(subset=['Critic_Score', 'Global_Sales']);

#create the plot
plt.scatter(x=df_cleaned['Critic_Score'], y=df_cleaned['Global_Sales'])

#give it a title
plt.title('Critic Score correlating to Global Sales')

#label the axis
plt.xlabel('Critic Score')
plt.ylabel('Global Sales')
plt.show()
```

```python
#you can kind of visually tell theres a little correlation between global sales and critic score
#BUT let us confirm the level of correlation using a Regression plot. (remember, omit any rows with Null)

# regression plot (make sure x and y are of same datatype), also color the scatter and regression line differently.
sns.regplot(x='Critic_Score', y='Global_Sales', data=df_cleaned[['Critic_Score', 'Global_Sales']].astype(float), scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#small correlation but not much visually. We will look at how to see HOW MUCH correlation there is.
```

```python
#numerically shows how much each coloumn correlates with the others.
df.corr() #the 3 main correlation techniques: Pearson, kendall, spearman
#df.corr() uses pearson by default.
#df.corr(method='pearson')
#research each correlation and when its best to use each.
```

```python
#0.24 pearson correlation between global sales and critic score, not much

#we can visualise all correlations in a correlation matrix.
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

#also label the axis.
#give it a title
plt.title('Correlation Matrix for Game Data')

#label the axis
plt.xlabel('Attributes')
plt.ylabel('Attributes')
plt.show()
```

```python
#we will now look at how Company correlates to the other columns (even though its non numeric)

#we can do this by giving a numerical UID to each company.

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'): #if columns datatype is object
        df_numerized[col_name] = df_numerized[col_name].astype('category') #set its type to category AND give it a specific number for each value
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr()
```

```python
#we can use something called unstacking to show which attributes have the highest correlation quickly.
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

#unstacking shows each attribute, then how it correlates to the other attributes, but separately.
corr_pairs
```

```python
#to show all rows in the above, would need to use the pd.set_option or smth, coverd it before.
#sort the correlations and pair them.
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#we can filter out which pairs we want to take.
#E.g. return all >0.5 correlations.
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
